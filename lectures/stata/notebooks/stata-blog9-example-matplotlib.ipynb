```python
import yfinance as yf
dowjones = yf.Ticker("^DJI")
aapl = yf.Ticker("AAPL")
msft = yf.Ticker("MSFT")
```

```python
data = dowjones.history(start="2010-01-01", end="2020-12-31")[['Close', 'Volume']]
```

```python
data
```

```python
data['Volume'] = data['Volume'] / 1000000
```

```python
data
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()

color = "g"
ax1.plot(data.index, data.Close, color)
ax1.set_ylabel('Closing Price', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_yticks(range(5000, 30001, 5000))

color = "b"
ax2 = ax1.twinx()
ax2.plot(data.index, data.Volume, color=color)
ax2.set_ylabel('Volume ($Millions)', color=color)
ax2.fill_between(data.index, data.Volume, color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yticks(range(0, 30+1, 5))

plt.title("Dow Jones Industrial Average (2010 - 2019)")
plt.savefig("stata-blog9-example-matplotlib.png", dpi=300)
```
